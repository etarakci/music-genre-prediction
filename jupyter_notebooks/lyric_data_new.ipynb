### Extract Data from Downloaded AZ Lyrics Database

```python
# AZ LYRICS
import glob
az_lyrics = []
path = "../final-data/azlyrics-scraper/*.csv"
for fname in glob.glob(path):
#     az_lyrics.append(pd.read_csv(fname, sep='delimiter')) #876 rows
#     az_lyrics.append(pd.read_csv(fname, error_bad_lines=False)) #760 rows
    n = 5
    az_lyrics.append(pd.read_csv(fname, usecols=range(n),lineterminator='\n',)) #868 rows
az_lyrics[0]
```

### Combine and Clean Columns Names

```python
# combine dataframes
az_data = pd.concat(az_lyrics)
az_data.rename(columns={"ARTIST_NAME":"artist_name","ARTIST_URL":"artist_url","SONG_NAME":"track_name","SONG_URL":"song_url","LYRICS":"lyrics"},inplace=True)
az_data
```
